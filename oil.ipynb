```python
from smart_choice.decisiontree import DecisionTree
from smart_choice.examples import oil_tree_example
from smart_choice.tornado_graph import tornado_graph
from smart_choice.value_sensitivity import ValueSensitivity

nodes = oil_tree_example()
tree = DecisionTree(nodes=nodes)
tree.evaluate()
tree.rollback()

sensitivities = {}
sensitivities['large-well'] = ValueSensitivity(
        decisiontree=tree,
        varname="oil_found",
        branch_name="large-well",
        values=(2500, 5000),
        single=True,
    )

sensitivities['drill cost'] =  ValueSensitivity(
        decisiontree=tree,
        varname="drill_decision",
        branch_name="drill",
        values=(450, 750),
        single=True,
    )

tornado_graph(sensitivities)


```
